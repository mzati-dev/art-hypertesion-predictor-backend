```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.feature_selection import SelectKBest, f_classif
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load and Clean Data

print("="*60)
print("STEP 1: LOADING AND CLEANING DATA")
print("="*60)

# Load the data
df = pd.read_csv('/content/sample_data/complete ART data.csv')
print(f"Original dataset shape: {df.shape}")

# Remove obvious outliers (like SYSTOLIC = 12)
df = df[df['SYSTOLIC'] >= 50].copy()
print(f"Shape after removing outliers: {df.shape}")

# Remove duplicate rows
df = df.drop_duplicates()
print(f"Shape after removing duplicates: {df.shape}")

# Check for missing values
print(f"\nMissing values:\n{df.isnull().sum()}")

# Display basic info
print("\nDataset Info:")
print(df.info())

print("\nFirst 5 rows:")
print(df.head())
```

```python
# Create Target Variable (Without Leakage)

print("\n" + "="*60)
print("STEP 2: CREATING TARGET VARIABLE")
print("="*60)

# Create hypertension target using standard criteria (ACC/AHA guidelines)
# Note: This uses BP values, but we will NOT use BP as features
df['HYPERTENSION'] = ((df['SYSTOLIC'] >= 130) | (df['DYSTOLIC'] >= 80)).astype(int)

print(f"Hypertension distribution:")
print(df['HYPERTENSION'].value_counts())
print(f"\nPercentage with Hypertension: {df['HYPERTENSION'].mean()*100:.2f}%")

# Visualize the target distribution
plt.figure(figsize=(6, 4))
df['HYPERTENSION'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Hypertension Distribution')
plt.xticks([0, 1], ['No Hypertension', 'Hypertension'], rotation=0)
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# CELL 4: Feature Engineering

print("\n" + "="*60)
print("STEP 3: FEATURE ENGINEERING")
print("="*60)

# Encode categorical variables
le_sex = LabelEncoder()
df['SEX_ENCODED'] = le_sex.fit_transform(df['SEX'])  # M=1, F=0
print("Sex encoding: M→1, F→0")

# Calculate pulse pressure (for EDA only, NOT for modeling)
df['PULSE_PRESSURE'] = df['SYSTOLIC'] - df['DYSTOLIC']

# Create BMI categories
df['BMI_CATEGORY'] = pd.cut(df['BODY MASS INDEX'],
                            bins=[0, 18.5, 25, 30, 100],
                            labels=['Underweight', 'Normal', 'Overweight', 'Obese'])

# Encode BMI categories
le_bmi = LabelEncoder()
df['BMI_CAT_ENCODED'] = le_bmi.fit_transform(df['BMI_CATEGORY'].astype(str))
print("\nBMI Categories distribution:")
print(df['BMI_CATEGORY'].value_counts())

# Create age groups
df['AGE_GROUP'] = pd.cut(df['AGE'],
                         bins=[0, 30, 40, 50, 60, 100],
                         labels=['<30', '30-40', '40-50', '50-60', '60+'])
le_age = LabelEncoder()
df['AGE_GROUP_ENCODED'] = le_age.fit_transform(df['AGE_GROUP'].astype(str))
print("\nAge Groups distribution:")
print(df['AGE_GROUP'].value_counts())

# Visualize relationships
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Age distribution by hypertension
df.boxplot(column='AGE', by='HYPERTENSION', ax=axes[0])
axes[0].set_title('Age by Hypertension Status')
axes[0].set_xlabel('Hypertension')

# BMI distribution by hypertension
df.boxplot(column='BODY MASS INDEX', by='HYPERTENSION', ax=axes[1])
axes[1].set_title('BMI by Hypertension Status')
axes[1].set_xlabel('Hypertension')

# Years on ART by hypertension
df.boxplot(column='YEARS ON ART', by='HYPERTENSION', ax=axes[2])
axes[2].set_title('Years on ART by Hypertension Status')
axes[2].set_xlabel('Hypertension')

plt.suptitle('')
plt.tight_layout()
plt.show()
```

```python
# CELL 5: Select Features (Excluding BP Measurements)


print("\n" + "="*60)
print("STEP 4: SELECTING FEATURES")
print("="*60)

# CRITICAL: EXCLUDE SYSTOLIC and DYSTOLIC to prevent leakage
feature_columns = [
    'AGE',                    # Demographics
    'SEX_ENCODED',            # Demographics
    'BODY MASS INDEX',        # Clinical measurement
    'YEARS ON ART',           # Treatment duration
    'BP HISTORY',             # Medical history (0/1)
    'EXERCISES',              # Lifestyle (0/1)
    'BMI_CAT_ENCODED',        # Categorized BMI
    'AGE_GROUP_ENCODED'       # Categorized age
]

# Add medication columns that show variability
med_columns = ['TENOFOVIR', 'LAMIVUDINE', 'DOLUTEGRAVIR', 'DARUNAVIR', 'ZIDOVUDINE', 'ABACAVIR']
for col in med_columns:
    if df[col].nunique() > 1:
        feature_columns.append(col)
        print(f"Added medication feature: {col}")
    else:
        print(f"Skipped {col} (no variability)")

print(f"\nFinal feature set ({len(feature_columns)} features):")
for i, col in enumerate(feature_columns, 1):
    print(f"  {i}. {col}")

# Check correlations among features (excluding target)
plt.figure(figsize=(12, 8))
correlation_matrix = df[feature_columns + ['HYPERTENSION']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
# Prepare Data for Modeling

print("\n" + "="*60)
print("STEP 5: PREPARING DATA FOR MODELING")
print("="*60)

X = df[feature_columns].copy()
y = df['HYPERTENSION'].copy()

# Handle any missing values
if X.isnull().sum().sum() > 0:
    X = X.fillna(X.mean())
    print("Filled missing values with mean")

# Split the data with stratification
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

print(f"Training set size: {len(X_train)}")
print(f"Test set size: {len(X_test)}")
print(f"\nTraining set hypertension rate: {y_train.mean()*100:.2f}%")
print(f"Test set hypertension rate: {y_test.mean()*100:.2f}%")

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Convert back to DataFrame for interpretability
X_train_scaled = pd.DataFrame(X_train_scaled, columns=feature_columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=feature_columns)

print("\n Data preparation complete")
print(f"Scaled training data shape: {X_train_scaled.shape}")
print(f"Scaled test data shape: {X_test_scaled.shape}")
```

```python
# Feature Selection to Reduce Overfitting

print("\n" + "="*60)
print("STEP 6: FEATURE SELECTION")
print("="*60)

# Use SelectKBest to identify most important features
k_features = min(10, len(feature_columns))
selector = SelectKBest(f_classif, k=k_features)
X_train_selected = selector.fit_transform(X_train_scaled, y_train)
X_test_selected = selector.transform(X_test_scaled)

# Get selected feature names
selected_mask = selector.get_support()
selected_features = [feature_columns[i] for i in range(len(feature_columns)) if selected_mask[i]]

print(f"Selected {len(selected_features)} most important features:")
for i, feature in enumerate(selected_features, 1):
    print(f"  {i}. {feature}")

# Get feature scores
feature_scores = pd.DataFrame({
    'feature': feature_columns,
    'score': selector.scores_,
    'selected': selected_mask
}).sort_values('score', ascending=False)

print("\nFeature Importance Scores:")
print(feature_scores)

# Visualize feature scores
plt.figure(figsize=(10, 6))
plt.barh(feature_scores['feature'], feature_scores['score'], color=['green' if s else 'gray' for s in feature_scores['selected']])
plt.xlabel('F-Score')
plt.title('Feature Importance (SelectKBest)')
plt.axvline(x=feature_scores['score'].iloc[k_features-1], color='red', linestyle='--', label=f'Cutoff for top {k_features}')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Setup Cross-Validation

print("\n" + "="*60)
print("STEP 7: SETTING UP CROSS-VALIDATION")
print("="*60)

# Setup cross-validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
print(f"Using {cv.n_splits}-fold Stratified Cross-Validation")

# Dictionary to store results
results = {}
```

```python
# Model 1 - Logistic Regression (with regularization)

print("\n" + "="*60)
print("MODEL 1: LOGISTIC REGRESSION (with regularization)")
print("="*60)

# Try different regularization strengths
C_values = [0.001, 0.01, 0.1, 1.0, 10.0]
cv_scores_lr = []

for C in C_values:
    lr = LogisticRegression(C=C, penalty='l2', random_state=42, max_iter=1000)
    scores = cross_val_score(lr, X_train_selected, y_train, cv=cv, scoring='roc_auc')
    cv_scores_lr.append(scores.mean())
    print(f"C={C}: CV ROC-AUC = {scores.mean():.4f} (±{scores.std():.4f})")

# Select best C
best_C = C_values[np.argmax(cv_scores_lr)]
best_lr = LogisticRegression(C=best_C, penalty='l2', random_state=42, max_iter=1000)
best_lr.fit(X_train_selected, y_train)

print(f"\n Best C: {best_C}")

# Predictions
y_pred_lr = best_lr.predict(X_test_selected)
y_pred_proba_lr = best_lr.predict_proba(X_test_selected)[:, 1]

# Evaluation
test_auc_lr = roc_auc_score(y_test, y_pred_proba_lr)
test_acc_lr = accuracy_score(y_test, y_pred_lr)

print(f"\nTest ROC-AUC: {test_auc_lr:.4f}")
print(f"Test Accuracy: {test_acc_lr:.4f}")

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred_lr, target_names=['No Hypertension', 'Hypertension']))

results['Logistic Regression'] = {
    'model': best_lr,
    'test_auc': test_auc_lr,
    'test_accuracy': test_acc_lr,
    'predictions': y_pred_lr,
    'probabilities': y_pred_proba_lr,
    'best_params': {'C': best_C}
}

# Plot C vs CV Score
plt.figure(figsize=(8, 5))
plt.plot(C_values, cv_scores_lr, 'bo-')
plt.xscale('log')
plt.xlabel('Regularization Strength (C)')
plt.ylabel('Cross-Validation ROC-AUC')
plt.title('Logistic Regression: C vs CV Score')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Random Forest (with regularization)

print("\n" + "="*60)
print("MODEL 2: RANDOM FOREST (with regularization)")
print("="*60)

# Define parameter grid
param_grid_rf = {
    'n_estimators': [50, 100],
    'max_depth': [3, 5, 7, None],
    'min_samples_split': [5, 10, 20],
    'min_samples_leaf': [2, 4]
}

# Grid search with cross-validation
rf = RandomForestClassifier(random_state=42, n_jobs=-1)
grid_rf = GridSearchCV(rf, param_grid_rf, cv=cv, scoring='roc_auc', n_jobs=-1, verbose=1)
grid_rf.fit(X_train_selected, y_train)

print(f"\n Best parameters: {grid_rf.best_params_}")
print(f"Best CV ROC-AUC: {grid_rf.best_score_:.4f}")

best_rf = grid_rf.best_estimator_

# Predictions
y_pred_rf = best_rf.predict(X_test_selected)
y_pred_proba_rf = best_rf.predict_proba(X_test_selected)[:, 1]

# Evaluation
test_auc_rf = roc_auc_score(y_test, y_pred_proba_rf)
test_acc_rf = accuracy_score(y_test, y_pred_rf)

print(f"\nTest ROC-AUC: {test_auc_rf:.4f}")
print(f"Test Accuracy: {test_acc_rf:.4f}")

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred_rf, target_names=['No Hypertension', 'Hypertension']))

# Feature importance
feature_importance = pd.DataFrame({
    'feature': selected_features,
    'importance': best_rf.feature_importances_
}).sort_values('importance', ascending=False)

print("\nFeature Importances:")
print(feature_importance)

# Plot feature importance
plt.figure(figsize=(10, 6))
plt.barh(feature_importance['feature'], feature_importance['importance'])
plt.xlabel('Importance')
plt.title('Random Forest Feature Importance')
plt.tight_layout()
plt.show()

results['Random Forest'] = {
    'model': best_rf,
    'test_auc': test_auc_rf,
    'test_accuracy': test_acc_rf,
    'predictions': y_pred_rf,
    'probabilities': y_pred_proba_rf,
    'best_params': grid_rf.best_params_,
    'feature_importance': feature_importance
}
```

```python
# CELL 11: Model 3 - Gradient Boosting (with regularization)
# ============================================

print("\n" + "="*60)
print("MODEL 3: GRADIENT BOOSTING (with regularization)")
print("="*60)

# Define parameter grid
param_grid_gb = {
    'n_estimators': [50, 100],
    'max_depth': [2, 3, 4],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.8, 1.0]
}

# Grid search with cross-validation
gb = GradientBoostingClassifier(random_state=42)
grid_gb = GridSearchCV(gb, param_grid_gb, cv=cv, scoring='roc_auc', n_jobs=-1, verbose=1)
grid_gb.fit(X_train_selected, y_train)

print(f"\n✅ Best parameters: {grid_gb.best_params_}")
print(f"Best CV ROC-AUC: {grid_gb.best_score_:.4f}")

best_gb = grid_gb.best_estimator_

# Predictions
y_pred_gb = best_gb.predict(X_test_selected)
y_pred_proba_gb = best_gb.predict_proba(X_test_selected)[:, 1]

# Evaluation
test_auc_gb = roc_auc_score(y_test, y_pred_proba_gb)
test_acc_gb = accuracy_score(y_test, y_pred_gb)

print(f"\nTest ROC-AUC: {test_auc_gb:.4f}")
print(f"Test Accuracy: {test_acc_gb:.4f}")

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred_gb, target_names=['No Hypertension', 'Hypertension']))

results['Gradient Boosting'] = {
    'model': best_gb,
    'test_auc': test_auc_gb,
    'test_accuracy': test_acc_gb,
    'predictions': y_pred_gb,
    'probabilities': y_pred_proba_gb,
    'best_params': grid_gb.best_params_
}
```

```python
# Compare Models and Check for Overfitting

print("\n" + "="*60)
print("STEP 8: MODEL COMPARISON")
print("="*60)

# Create comparison dataframe
comparison_df = pd.DataFrame({
    'Model': list(results.keys()),
    'Test Accuracy': [results[m]['test_accuracy'] for m in results.keys()],
    'Test ROC-AUC': [results[m]['test_auc'] for m in results.keys()]
})

print("\nModel Performance Comparison:")
print(comparison_df.to_string(index=False))

# Identify best model
best_model_name = max(results.keys(), key=lambda x: results[x]['test_auc'])
best_model = results[best_model_name]['model']
print(f"\n BEST MODEL: {best_model_name}")
print(f"   ROC-AUC = {results[best_model_name]['test_auc']:.4f}")
print(f"   Accuracy = {results[best_model_name]['test_accuracy']:.4f}")

# Visual comparison
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Accuracy comparison
axes[0].bar(comparison_df['Model'], comparison_df['Test Accuracy'], color=['blue', 'green', 'orange'])
axes[0].set_ylabel('Accuracy')
axes[0].set_title('Model Accuracy Comparison')
axes[0].set_ylim([0, 1])
for i, v in enumerate(comparison_df['Test Accuracy']):
    axes[0].text(i, v + 0.01, f'{v:.3f}', ha='center')

# AUC comparison
axes[1].bar(comparison_df['Model'], comparison_df['Test ROC-AUC'], color=['blue', 'green', 'orange'])
axes[1].set_ylabel('ROC-AUC')
axes[1].set_title('Model ROC-AUC Comparison')
axes[1].set_ylim([0, 1])
for i, v in enumerate(comparison_df['Test ROC-AUC']):
    axes[1].text(i, v + 0.01, f'{v:.3f}', ha='center')

plt.tight_layout()
plt.show()
```

```python
# Confusion Matrix for Best Model

print("\n" + "="*60)
print("STEP 9: CONFUSION MATRIX - BEST MODEL")
print("="*60)

cm = confusion_matrix(y_test, results[best_model_name]['predictions'])

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Hypertension', 'Hypertension'],
            yticklabels=['No Hypertension', 'Hypertension'])
plt.title(f'Confusion Matrix - {best_model_name}')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.tight_layout()
plt.show()

# Calculate metrics
tn, fp, fn, tp = cm.ravel()
sensitivity = tp / (tp + fn)
specificity = tn / (tn + fp)
ppv = tp / (tp + fp)
npv = tn / (tn + fn)

print(f"\nDetailed Metrics for {best_model_name}:")
print(f"Sensitivity (Recall): {sensitivity:.4f}  (Ability to detect hypertension)")
print(f"Specificity: {specificity:.4f}  (Ability to identify non-hypertension)")
print(f"Positive Predictive Value: {ppv:.4f}  (Precision)")
print(f"Negative Predictive Value: {npv:.4f}")
print(f"False Positive Rate: {fp/(fp+tn):.4f}")
print(f"False Negative Rate: {fn/(fn+tp):.4f}")
```

```python
# ROC Curves for All Models

print("\n" + "="*60)
print("STEP 10: ROC CURVES")
print("="*60)

plt.figure(figsize=(10, 8))

for model_name in results.keys():
    fpr, tpr, _ = roc_curve(y_test, results[model_name]['probabilities'])
    auc = results[model_name]['test_auc']
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {auc:.3f})', linewidth=2)

plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier', alpha=0.5)
plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate', fontsize=12)
plt.title('ROC Curves Comparison', fontsize=14)
plt.legend(loc='lower right', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Overfitting Check
print("\n" + "="*60)
print("STEP 11: OVERFITTING CHECK")
print("="*60)

# Calculate training performance for each model
overfitting_results = []

for model_name in results.keys():
    model = results[model_name]['model']

    # Get training predictions
    if model_name == 'Logistic Regression':
        train_pred = model.predict(X_train_selected)
    else:
        train_pred = model.predict(X_train_selected)

    train_accuracy = accuracy_score(y_train, train_pred)
    test_accuracy = results[model_name]['test_accuracy']
    performance_gap = train_accuracy - test_accuracy

    overfitting_results.append({
        'Model': model_name,
        'Train Accuracy': train_accuracy,
        'Test Accuracy': test_accuracy,
        'Gap': performance_gap,
        'Status': ' Good' if performance_gap < 0.1 else (' Warning' if performance_gap < 0.2 else ' Overfitting')
    })

overfitting_df = pd.DataFrame(overfitting_results)
print("\nOverfitting Analysis:")
print(overfitting_df.to_string(index=False))

# Visualize overfitting
plt.figure(figsize=(10, 6))
x = np.arange(len(overfitting_results))
width = 0.35

train_acc = [r['Train Accuracy'] for r in overfitting_results]
test_acc = [r['Test Accuracy'] for r in overfitting_results]

plt.bar(x - width/2, train_acc, width, label='Training Accuracy', color='skyblue')
plt.bar(x + width/2, test_acc, width, label='Test Accuracy', color='salmon')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Training vs Test Accuracy (Overfitting Check)')
plt.xticks(x, [r['Model'] for r in overfitting_results], rotation=15)
plt.legend()
plt.ylim([0, 1])

# Add gap labels
for i, gap in enumerate([r['Gap'] for r in overfitting_results]):
    plt.text(i, max(train_acc[i], test_acc[i]) + 0.02, f'Gap: {gap:.3f}', ha='center')

plt.tight_layout()
plt.show()
```

```python
# Create Prediction Function


print("\n" + "="*60)
print("STEP 12: CREATING PREDICTION FUNCTION")
print("="*60)

def predict_hypertension_risk(patient_data, model, selector, scaler, selected_features, feature_columns):
    """
    Predict hypertension risk for a new patient

    Parameters:
    patient_data: dict with patient information
    model: trained model
    selector: fitted feature selector
    scaler: fitted scaler
    selected_features: list of selected feature names
    feature_columns: list of all feature columns

    Returns:
    dict with prediction results
    """
    # Create DataFrame with all possible features
    patient_df = pd.DataFrame([patient_data])

    # Ensure all features are present
    for col in feature_columns:
        if col not in patient_df.columns:
            patient_df[col] = 0

    # Reorder columns
    patient_df = patient_df[feature_columns]

    # Scale
    patient_scaled = scaler.transform(patient_df)

    # Apply feature selection
    patient_selected = selector.transform(patient_scaled)

    # Predict
    prediction = model.predict(patient_selected)[0]
    probability = model.predict_proba(patient_selected)[0][1]

    # Risk category
    if probability < 0.3:
        risk_category = "Low Risk"
        color = 'green'
    elif probability < 0.6:
        risk_category = "Moderate Risk"
        color = 'orange'
    else:
        risk_category = "High Risk"
        color = 'red'

    return {
        'prediction': 'Hypertension' if prediction == 1 else 'No Hypertension',
        'probability': probability,
        'risk_category': risk_category,
        'risk_score': f"{probability:.1%}",
        'color': color
    }

print("Prediction function created successfully")

# Test with example patients
print("\n" + "="*60)
print("TESTING PREDICTION FUNCTION")
print("="*60)

# Example 1: High-risk patient
example_patient_high = {
    'AGE': 65,
    'SEX_ENCODED': 0,  # Female
    'BODY MASS INDEX': 32.5,
    'YEARS ON ART': 15,
    'BP HISTORY': 1,    # Has history
    'EXERCISES': 0,     # Doesn't exercise
    'BMI_CAT_ENCODED': 3,  # Obese
    'AGE_GROUP_ENCODED': 4,  # 60+
    'TENOFOVIR': 1,
    'LAMIVUDINE': 1,
    'DOLUTEGRAVIR': 1
}

# Example 2: Moderate-risk patient
example_patient_moderate = {
    'AGE': 45,
    'SEX_ENCODED': 1,  # Male
    'BODY MASS INDEX': 26.0,
    'YEARS ON ART': 8,
    'BP HISTORY': 0,    # No history
    'EXERCISES': 1,     # Exercises
    'BMI_CAT_ENCODED': 2,  # Overweight
    'AGE_GROUP_ENCODED': 2,  # 40-50
    'TENOFOVIR': 1,
    'LAMIVUDINE': 1,
    'DOLUTEGRAVIR': 1
}

# Example 3: Low-risk patient
example_patient_low = {
    'AGE': 30,
    'SEX_ENCODED': 0,  # Female
    'BODY MASS INDEX': 22.0,
    'YEARS ON ART': 3,
    'BP HISTORY': 0,    # No history
    'EXERCISES': 1,     # Exercises
    'BMI_CAT_ENCODED': 1,  # Normal
    'AGE_GROUP_ENCODED': 1,  # 30-40
    'TENOFOVIR': 1,
    'LAMIVUDINE': 1,
    'DOLUTEGRAVIR': 1
}

# Test all examples
for i, (name, patient) in enumerate([
    ("High-Risk", example_patient_high),
    ("Moderate-Risk", example_patient_moderate),
    ("Low-Risk", example_patient_low)
]):
    result = predict_hypertension_risk(
        patient,
        best_model,
        selector,
        scaler,
        selected_features,
        feature_columns
    )

    print(f"\n{'-'*40}")
    print(f"{name} Patient (Age: {patient['AGE']}, BMI: {patient['BODY MASS INDEX']})")
    print(f"Prediction: {result['prediction']}")
    print(f"Risk Score: {result['risk_score']}")
    print(f"Risk Category: {result['risk_category']}")
```

```python
# Generate Clinical Decision Support Tool

print("\n" + "="*60)
print("STEP 15: CLINICAL DECISION SUPPORT TOOL")
print("="*60)

def get_clinical_recommendation(probability, age, bmi, years_on_art, bp_history):
    """
    Generate clinical recommendations based on risk score
    """
    recommendations = []

    if probability >= 0.6:
        recommendations.append("HIGH RISK: Initiate or intensify antihypertensive therapy")
        recommendations.append("Monitor BP weekly")
        recommendations.append(" Consider cardiology referral")
        recommendations.append(" Review ART regimen for potential interactions")
    elif probability >= 0.3:
        recommendations.append(" MODERATE RISK: Lifestyle modifications recommended")
        recommendations.append(" Monitor BP monthly")
        recommendations.append("Consider starting low-dose antihypertensive if other risk factors present")
    else:
        recommendations.append("LOW RISK: Continue current management")
        recommendations.append("Monitor BP every 3-6 months")
        recommendations.append("Reinforce healthy lifestyle habits")

    # Additional risk-based recommendations
    if age > 60:
        recommendations.append(" Age >60: Annual cardiovascular assessment recommended")

    if bmi > 30:
        recommendations.append(" Obese: Refer to nutritionist for weight management")
    elif bmi > 25:
        recommendations.append(" Overweight: Counsel on weight management")

    if years_on_art > 10:
        recommendations.append(" Long-term ART: Monitor for metabolic complications")

    if bp_history == 1:
        recommendations.append(" History of BP issues: Continue close monitoring")

    return recommendations

# Test the clinical tool
for name, patient in [
    ("High-Risk", example_patient_high),
    ("Moderate-Risk", example_patient_moderate),
    ("Low-Risk", example_patient_low)
]:
    result = predict_hypertension_risk(
        patient,
        loaded_model,
        loaded_selector,
        loaded_scaler,
        loaded_features,
        loaded_all_features
    )

    print(f"\n{'='*50}")
    print(f"PATIENT: {name}")
    print(f"Age: {patient['AGE']}, BMI: {patient['BODY MASS INDEX']}, Years on ART: {patient['YEARS ON ART']}")
    print(f"Risk Score: {result['risk_score']} - {result['risk_category']}")
    print(f"{'='*50}")

    recommendations = get_clinical_recommendation(
        result['probability'],
        patient['AGE'],
        patient['BODY MASS INDEX'],
        patient['YEARS ON ART'],
        patient['BP HISTORY']
    )

    for rec in recommendations:
        print(rec)
```
